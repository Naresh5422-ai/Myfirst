Python has excellent libraries for data visualization. A combination of Pandas, numpy and matplotlib can help in creating in nearly all types of visualizations charts.

#### Creating Charts

We use numpy library to create require number to be mapped for creating the charts and the pyplot  method in matpltlib to drawn the actual chart

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

```python
x = np.arange(0,10)
y = x ^ 2
#print(y)
# Now we create simple plot
plt.plot(x,y)
```

### Labeling the Axes

We can apply labels to the axes as well as a title for the chart using appropriate methods from the library as shown below.

```python
# Labeling the axes using xlabel, ylabel, title methods
plt.xlabel("X Values")
plt.ylabel("y Values")
plt.title("initial Graph")
plt.plot(x,y)
plt.show()
```

### Formating Line type and color

Apply the style as well as color to the graph

```python
# formating the line color
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Second Graph")
plt.plot(x,y, 'r')

# Formating the line type
plt.xlabel("X-axis")
plt.ylabel("Y-Axis")
plt.title("Show Line type")
plt.plot(x,y,'>')

# Now we have to save the graph using savefig() method
plt.savefig("First_Graph.pdf", format = 'pdf')
plt.show()
```

# Chart Styling

The charts created in python can have further styling by using some appropriate methods from the libraries used for charting. In this, we will see the implementation of Annotation, legends and chart background.

### Adding Annotations

Many times, we need to annotate the chart by highlighting the specific locations of the chart. In the below we indicate the sharp change in values in the chart by adding annotations at those points.

```python
z = x ^ 3
t = x ^ 4
plt.title("Graph drawing")
plt.xlabel("X-Axis")
plt.ylabel("Y-axis")
plt.plot(x,y)

# Annotate
plt.annotate(xy = [2,1], s = "Second entry")
plt.annotate(xy = [4,6], s = "Third entry")
```

### Adding Legends

We sometimes need a chart with multiple lines being plotted. Use of legend represents the meaning associated with each line. In the below chart we have 3 lines with appropriate legends.

```python
plt.xlabel("X-Axis")
plt.ylabel("Y-axis")
plt.title("Adding Legends")

# Annotate
plt.annotate(xy = [2,1], s = "Second Entry")
plt.annotate(xy = [4,6], s = "Third Entry")


# Addding Legends
plt.plot(x,y)
plt.plot(x, z)
plt.plot(x,t)
plt.legend(["First","Second","Third"], loc = 2)
plt.show()
```

### Chart Presentation Style

We can modify the presentation style of the chart by using different methods from the style package

```python
# Add labeling
plt.xlabel("X_Axis")
plt.ylabel("Y_Axis")
plt.title("Chart presentation style")
plt.plot(x,y)

# Annotate
plt.annotate(xy = [2,1], s = "Second entry")
plt.annotate(xy = [4,3], s = "Third Entry")

# Adding legends
plt.plot(x,z)
plt.plot(x,t)
plt.legend(["Race1","race2","Race3"], loc = 2)

# Create the background of chart
plt.style.use('fast')
plt.plot(x,z)
plt.show()
```
